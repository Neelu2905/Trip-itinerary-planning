```python
import pandas as pd
import matplotlib.pyplot as plt
import textwrap

# Load the Excel file
file_path = "/content/Data Record - 2,3.xlsx"
xls = pd.ExcelFile(file_path)

# Read the first sheet (modify if needed)
sheet_name = xls.sheet_names[4]  # Get the sheet name as city name
df = pd.read_excel(file_path, sheet_name=sheet_name)

# Ensure correct column names (trim spaces)
df.columns = df.columns.str.strip()

# Select relevant columns
df_filtered = df[["CONDITION", "Time of Computation (seconds)"]].dropna()

# Read Cost Budget and Time Budget from G2 and H2
cost_budget = df.iloc[0, 6]  # G2 -> Row index 0, Column index 6
time_budget = df.iloc[0, 7]  # H2 -> Row index 0, Column index 7

# Wrap long labels for x-axis
wrapped_labels = [textwrap.fill(label, width=15) for label in df_filtered["CONDITION"]]

# Plot line chart
plt.figure(figsize=(8, 5))
plt.plot(df_filtered["CONDITION"], df_filtered["Time of Computation (seconds)"], marker='o', linestyle='-', color='b')

# Set labels and title
plt.xlabel("Condition")
plt.ylabel("Time of Computation (seconds)")
plt.title("Computation Time under Different Conditions")
plt.xticks(df_filtered["CONDITION"], wrapped_labels, ha='center', multialignment='center')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Adjust y-axis to create space for text box
y_max = df_filtered["Time of Computation (seconds)"].max()
plt.ylim(0, y_max * 1.3)  # Add 30% extra space

# Add text inside the plot (at top-right but within the graph)
text_x = 0.7  # Adjust horizontal position (0 = left, 1 = right)
text_y = 0.95  # Adjust vertical position (closer to 1 = top)
text_str = f"City: {sheet_name}\nCost Budget: {cost_budget}\nTime Budget: {time_budget}"
plt.text(text_x, text_y, text_str, transform=plt.gca().transAxes,
         fontsize=10, verticalalignment='top', horizontalalignment='left',
         bbox=dict(facecolor='white', alpha=0.8, edgecolor='black'))

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the Excel file
file_path = "/content/Data Record - 1(1).xlsx"
xls = pd.ExcelFile(file_path)

# Read the first sheet (modify if needed)
sheet_name = xls.sheet_names[3]  # City name as title
df = pd.read_excel(file_path, sheet_name=sheet_name)

# Ensure correct column names (trim spaces)
df.columns = df.columns.str.strip()

# Select relevant columns
df_filtered = df[["Cost Budget", "Travel Time", "Visit Time"]].dropna()

# Read Time Budget from F2 (Column F)
time_budget = df.iloc[0, 5]  # Column index 5 (F2)

# Generate equidistant x positions
x_pos = np.arange(len(df_filtered))

# Define bar width
bar_width = 0.5

# Create the plot
plt.figure(figsize=(8, 5))

# Plot Visit Time (bottom part of the bar) - Black
plt.bar(x_pos, df_filtered["Visit Time"],
        color="#A9A9A9", label="Visit duration",edgecolor="black", width=bar_width)

# Plot Travel Time (stacked on top of Visit Time) - Gray
plt.bar(x_pos, df_filtered["Travel Time"],
        color="#FFFFFF", label="Travel time",edgecolor="black", width=bar_width,
        bottom=df_filtered["Visit Time"])

# Set labels and title
plt.xlabel("Cost Budget")
plt.ylabel("Time (minutes)")
plt.title(sheet_name)  # City name as title
plt.xticks(x_pos, df_filtered["Cost Budget"])  # Set correct x-axis labels
plt.ylim(0, max(df_filtered["Travel Time"] + df_filtered["Visit Time"]) * 1.2)  # Add extra space

# Move legend to the left top
plt.legend(loc="upper left", frameon=True)

# Add text on the top-right inside the plot
text_x = 0.95  # Adjust horizontal position
text_y = 0.95  # Adjust vertical position
text_str = f"Time Budget: {time_budget}"
plt.text(text_x, text_y, text_str, transform=plt.gca().transAxes,
         fontsize=10, verticalalignment='top', horizontalalignment='right',
         bbox=dict(facecolor='white', alpha=0.8, edgecolor='black'))

# Show the plot
plt.show()
```
